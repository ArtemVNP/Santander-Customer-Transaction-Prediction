```python
import numpy as np
import pandas as pd
import os, gc
from glob import glob
from tqdm import tqdm

import sys
sys.path.append(f'/home/{os.environ.get("USER")}/PythonLibrary')

import EDA
```

```python
tr = pd.read_csv('../input/train.csv.zip')
te = pd.read_csv('../input/test.csv.zip')
```

```python
def fe(df):
    PREF = 'uq'
    feature = pd.DataFrame(index=df.index)
    
    for c in tqdm(df.columns):
        di = df[c].value_counts().to_dict()
        feature[f'{PREF}_{c}'] = (df[c].map(di)==1)*1
    
    feature[f'{PREF}_sum'] = feature.sum(1)
    
    return feature
```

```python
tr_fe = fe(tr.drop(['ID_code', 'target'], axis=1))
te_fe = fe(te.drop(['ID_code'], axis=1))
```
